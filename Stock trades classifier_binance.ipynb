```python

import plotly.graph_objects as go
from datetime import datetime
import stock_marker_binance
```

```python
#The goal of this program is to mark the candlesticks with one of the following marks: Buy, Sell, Hold, Wait.
```

```python
SYMBOL = 'ETHUSDT'
INTERVAL = '1h'
df=stock_marker_binance.import_stock_data(SYMBOL, INTERVAL, limit=1000)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.rename(
          columns={0:'timestamp', 1:'open', 2:'high', 3:'low', 4:'close',5:'volume', 6:'close_time', 7:'quote_asset_volume',
                8:'num_trades', 9:'taker_buy_base_asset_volume',10:'taker_buy_quote_asset_volume', 11:'unused_field'},
         inplace=True)
```

```python
df.drop(columns=['taker_buy_base_asset_volume', 'taker_buy_quote_asset_volume', 'unused_field'], inplace=True)
```

```python
df.head()
```

```python
df_result = stock_marker_binance.adding_MAs(df, 'close')
```

```python
df_result = stock_marker_binance.marking_buy_sell_actions(df_result)
```

```python
df_result = stock_marker_binance.marking_adjacent_actions(df_result)
```

```python
df_result
```

```python
df_result = stock_marker_binance.marking_hold_wait_actions(df_result)
```

```python
df_result = stock_marker_binance.renaming_cols(df_result, SYMBOL + "_", ['timestamp', 'action','open' ,'close', 'low', 'high'])
```

```python
df_result.head()
```

```python
df_sell = df_result.loc[df_result['action'] == 'S' ]
df_sell['close'] = df_sell['close'].apply(lambda x: x*1.1)
# df_sell.loc[:,'close'] = df_sell['close'] * 1.1
```

```python
df_buy = df_result.loc[df_result['action'] == 'B' ]
df_buy['close'] = df_buy['close'].apply(lambda x: x*0.9)
# df_buy.loc[:,'close'] = df_buy['close'] * 0.9
```

```python
df_result.tail()
```

```python
fig2 = go.Figure(data=[go.Candlestick(x=df_result['timestamp'],
                open=df_result['open'],
                high=df_result['high'],
                low=df_result['low'],
                close=df_result['close'])])
```

```python
for date, price in zip(df_buy['timestamp'], df_buy['close'] ):
    fig2.add_annotation(x=date, y=price,
                text="B",
                showarrow=False,
                arrowhead=1,
                bordercolor="#00ff00",
                borderwidth=2,
                borderpad=4,
                bgcolor="#00ff00",)
```

```python
for date, price in zip(df_sell['timestamp'], df_sell['close']):
    fig2.add_annotation(x=date, y=price,
                text="S",
                showarrow=False,
                arrowhead=1,
                bordercolor="#ff0000",
                borderwidth=2,
                borderpad=4,
                bgcolor="#ff0000",)
```

```python
#fig2.update_layout(yaxis_range=[0,500])
fig2.show()
```
